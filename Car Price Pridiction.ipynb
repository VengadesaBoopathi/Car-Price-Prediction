Car Price Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
data = pd.read_csv('Data/CarPrice.csv')
```

```python
data.head()
```

```python
data.shape
```

Data PreProcessing

```python
data.describe()
```

```python
data.info()
```

```python
data['doornumber']=data['doornumber'].replace('two',2)
data['doornumber']=data['doornumber'].replace('four',4)
data['doornumber'] = data['doornumber'].astype(np.int64)
```

```python
data['cylindernumber']=data['cylindernumber'].replace('four',4)
data['cylindernumber']=data['cylindernumber'].replace('six',6 )
data['cylindernumber']=data['cylindernumber'].replace('five',5 )
data['cylindernumber']=data['cylindernumber'].replace('three',3 )
data['cylindernumber']=data['cylindernumber'].replace('twelve',12 )
data['cylindernumber']=data['cylindernumber'].replace('two',2 )
data['cylindernumber']=data['cylindernumber'].replace('eight',8 )
data['cylindernumber']=data['cylindernumber'].astype(np.int64)
```

```python
fueltype = pd.get_dummies(data['fueltype'],drop_first=True)
aspiration = pd.get_dummies(data['aspiration'],drop_first=True)
carbody = pd.get_dummies(data['carbody'],drop_first=True)
drivewheel = pd.get_dummies(data['drivewheel'],drop_first=True)
enginelocation = pd.get_dummies(data['enginelocation'],drop_first=True)
enginetype = pd.get_dummies(data['enginetype'],drop_first=True)
fuelsystem = pd.get_dummies(data['fuelsystem'],drop_first=True)
```

```python
data = pd.concat([data,fueltype,aspiration,carbody,drivewheel,enginelocation,enginetype,fuelsystem],axis=1)
data.head()
```

```python
data.head()
```

```python
data['gas'] = data['gas'].astype(np.int64)
data['turbo'] = data['turbo'].astype(np.int64)
data['hardtop'] = data['hardtop'].astype(np.int64)
data['hatchback'] = data['hatchback'].astype(np.int64)
data['sedan'] = data['sedan'].astype(np.int64)
data['wagon'] = data['wagon'].astype(np.int64)
data['fwd'] = data['fwd'].astype(np.int64)
data['rwd'] = data['rwd'].astype(np.int64)
data['rear'] = data['rear'].astype(np.int64)
data['dohcv'] = data['dohcv'].astype(np.int64)
data['l'] = data['l'].astype(np.int64)
data['ohc'] = data['ohc'].astype(np.int64)
data['ohcf'] = data['ohcf'].astype(np.int64)
data['ohcv'] = data['ohcv'].astype(np.int64)
data['rotor'] = data['rotor'].astype(np.int64)
data['2bbl'] = data['2bbl'].astype(np.int64)
data['4bbl'] = data['4bbl'].astype(np.int64)
data['idi'] = data['idi'].astype(np.int64)
data['mfi'] = data['mfi'].astype(np.int64)
data['mpfi'] = data['mpfi'].astype(np.int64)
data['spdi'] = data['spdi'].astype(np.int64)
data['spfi'] = data['spfi'].astype(np.int64)
```

```python
data.drop(['fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem','CarName'],axis=1,inplace=True)
```

```python
data.info()
```

Dividing the entire dataset into training and testing sets

```python
X = data.drop('price',axis=1)
y = data.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=101)
```

Fitting the Linear Regression model on the training set

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
predictions
```


Model Evaluation

```python
print('Mean Absolute Error:',mean_absolute_error(y_test,predictions))
print('Mean Squared Error:',mean_squared_error(y_test,predictions))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predictions)))
```
